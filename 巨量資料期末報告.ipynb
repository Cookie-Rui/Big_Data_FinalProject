```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA = pd.read_csv('/content/taiwan population data.csv')
```

```python
DATA
```

```python
DATA['年別(西元)'] = np.arange(1905, 2022)
```

```python
DATA
```

```python
DATA.dropna().describe()
```

```python
pop_pivot_table = pd.DataFrame({'population':DATA['人口數'].values}, index = np.arange(1905, 2022))
```

```python
pop_pivot_table
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
pop_pivot_table.plot(ax=ax)
ax.grid(True, linestyle='--')
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2e'))
ax.legend(fontsize = 25)
plt.xlabel('year', size = 20)
plt.ylabel('population', size = 20)
plt.title('The overall population change in Taiwan from 1905 to 2021', size=25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
gender_pivot_table = pd.DataFrame({'Male':DATA['男'].values, 'Female':DATA['女'].values}, index = np.arange(1905, 2022))
```

```python
gender_pivot_table
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
gender_pivot_table.plot(ax=ax)
ax.grid(True, linestyle='--')
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2e'))
ax.legend(fontsize = 25)
plt.xlabel('year', size = 20)
plt.ylabel('population', size = 20)
plt.title('The population change between male and female from 1905 to 2021', size=25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
birth_death_pivot_table = pd.DataFrame({'Birth':DATA['出生'].values, 'Death':DATA['死亡'].values}, index = np.arange(1905, 2022))
```

```python
birth_death_pivot_table
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
birth_death_pivot_table.plot(ax=ax)
ax.grid(True, linestyle='--')
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2e'))
ax.legend(fontsize = 25, loc='upper left')
plt.xlabel('year', size = 20)
plt.ylabel('population', size = 20)
plt.title('The Birth and Death change in Taiwan from 1905 to 2021', size=25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
marriage_pivot_table = pd.DataFrame({'Married':DATA['結婚對數'].values, 'Divorced':DATA['離婚對數'].values}, index = np.arange(1905, 2022))
```

```python
marriage_pivot_table
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
marriage_pivot_table.plot(ax=ax)
ax.grid(True, linestyle='--')
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2e'))
ax.legend(fontsize = 25, loc='upper left')
plt.xlabel('year', size = 20)
plt.ylabel('number', size = 20)
plt.title('The Married and Divorced change in Taiwan from 1905 to 2021', size=25)
plt.xticks(size = 15)
plt.yticks(size = 15)
```

```python
the_gap = marriage_pivot_table['Married'] - marriage_pivot_table['Divorced']
marriage_gap = (the_gap - np.mean(the_gap)) / np.std(the_gap)
```

```python
marriage_gap
```

```python
fig, ax = plt.subplots(figsize=(17, 8))
marriage_gap.plot(ax=ax)
ax.grid(True, linestyle='--')
ax.yaxis.set_major_formatter(plt.FormatStrFormatter('%.2e'))
plt.legend()
ax.get_legend().set_visible(False)
plt.xlim(1898, 2026)
plt.xlabel('year', size = 20)
plt.ylabel('number', size = 20)
plt.title('The change of \"standardized\" difference between married and divorced', size=25)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.hlines(0,1898,2026,color="red")
ax.text(1900, 0.05, "z-score=0-->mean(married - divorced)", size=20, color='gray')
plt.arrow(1900, -0.05, 0, -0.85, width=1, head_length=0.25)
plt.arrow(1900, 0.25, 0, 0.85, width=1, head_length=0.25)
ax.text(1902, 0.7, "above the mean", size=22, color='gray')
ax.text(1902, -0.7, "below the mean", size=22, color='gray')
```
